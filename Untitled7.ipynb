```python
def reverseInParentheses(inputString):
    new_str = []
    count = 0
    for i in range(len(inputString)):
        if inputString[i] == ')':
            for y in range(i - 1, 0, -1):
                print(inputString[y])
                if inputString[y] == '(':
                    break
                new_str.append(inputString[y])
        elif inputString[i] == '(':
            for x in range(0, i):
                if inputString[i] == ')':
                    continue
                new_str.append(inputString[x])
    return new_str
```
